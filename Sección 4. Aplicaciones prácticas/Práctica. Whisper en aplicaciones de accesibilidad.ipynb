## Generando la transcripción en audio

A continuación llamamos a la API para transcribir un mensaje de texto:

```python
texto = """
Tengo miedo a perder la maravilla
de tus ojos de estatua y el acento
que me pone de noche en la mejilla
la solitaria rosa de tu aliento.
Tengo pena de ser en esta orilla
tronco sin ramas, y lo que más siento
es no tener la flor, pulpa o arcilla,
para el gusano de mi sufrimiento.
"""
```

```python
archivo_audio = "poema_Lorca.mp3"

resultado = client.audio.speech.create(
  model="tts-1",
  voice="alloy",
  input=texto
)
```

```python
resultado
```

Con esta llamada a la API ya hemos generado el audio. Comprobemos que la llamada ha sido correcta:

```python
resultado.stream_to_file(archivo_audio)
```

Con estas pocas líneas de código hemos sido capaces de generar un audio a partir de las dos primeras estrofas de un poema. Veamos además que esta aplicación nos permite adaptar estos audios, por ejemplo, podemos usar otra voz:

```python
archivo_audio = "poema_Lorca_onyx.mp3"

resultado = client.audio.speech.create(
  model="tts-1",
  voice="onyx",
  input=texto
)
resultado.stream_to_file(archivo_audio)
```

También podemos generar el archivo en alta definición para una mayor nitidez:

```python
archivo_audio = "poema_Lorca_onyx_hd.mp3"

resultado = client.audio.speech.create(
  model="tts-1-hd",
  voice="onyx",
  input=texto
)
resultado.stream_to_file(archivo_audio)
```

Por último también podemos jugar con la velocidad con la que queremos que "el locutor" hable:

```python
archivo_audio = "poema_Lorca_onyx_hd_rapido.mp3"

resultado = client.audio.speech.create(
  model="tts-1-hd",
  voice="onyx",
  input=texto,
  speed=3.0
)
resultado.stream_to_file(archivo_audio)
```

```python
archivo_audio = "poema_Lorca_onyx_hd_lento.mp3"

resultado = client.audio.speech.create(
  model="tts-1-hd",
  voice="onyx",
  input=texto,
  speed=0.25
)
resultado.stream_to_file(archivo_audio)
```
